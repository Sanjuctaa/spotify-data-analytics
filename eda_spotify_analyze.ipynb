# Top 5 popular artists

```python
top_5_artist= df.groupby('artist(s)_name').count().sort_values(by= 'track_name',ascending= False)['track_name'][:5]
top_5_artist                                                                                                     
```

```python
top_5_artist.plot.barh()
plt.show()
```

```python
#top5 energetic songs

top_5_energetic_song = df[['track_name','energy_%']].sort_values(by= 'energy_%', ascending= False).head(5)
top_5_energetic_song
```

```python
sns.barplot(x='energy_%',y='track_name',data= top_5_energetic_song)
plt.title('Top 5 energetic tracks')
```

```python
#artist with most danceable songs

top_5_danceable_song = df[['track_name','danceability_%']].sort_values(by= 'danceability_%', ascending= False).head(5)
top_5_danceable_song
```

```python
sns.barplot(x='danceability_%',y= 'track_name', data= top_5_danceable_song)
plt.title('top_5_danceable_song')
```

```python
#top 10 instrumental tracks

top_5_instru_song = df[['track_name','instrumentalness_%']].sort_values(by= 'instrumentalness_%', ascending= False).head(5)
top_5_instru_song
```

```python
plt.pie(x= 'instrumentalness_%',data= top_5_instru_song ,autopct='%1.2f%%',labels = top_5_instru_song.track_name )
plt.show()
```
